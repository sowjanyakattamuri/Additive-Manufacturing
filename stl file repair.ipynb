```python
df = pd.read_excel('/content/drive/MyDrive/Vert.xlsx', names=['x', 'y', 'z'])
```

```python
df=df.dropna()
```

```python
n = len(df)
```

```python
l = []
for i in range(n):
  l.append([df.iloc[i]['x'], df.iloc[i]['y'], df.iloc[i]['z']])
```

```python
l2 = []
N = len(l)
for i in range(N//3):
  l2.append([l[3*i], l[3*i+1]])
  l2.append([l[3*i+1], l[3*i+2]])
  l2.append([l[3*i+2], l[3*i]])
```

```python
counts = [0 for i in range(len(l2))]
for i in range(len(l2)):
  for j in range(len(l2)):
    if l2[i] == l2[j] or (l2[i][0]==l2[j][1] and l2[i][1]==l2[j][0]):
      counts[i]+=1
```

```python
from collections import Counter

c = Counter(counts)
```

```python
Bad_edges = []
for i in range(len(counts)):
  if counts[i]==1:
    Bad_edges.append(l2[i])
```
